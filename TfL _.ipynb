###**Insight 1:** What is the average journey duration by days of the week?



```python
# Calculating Journey Duration
tfl_df['JourneyDuration'] = tfl_df['EXTimeHHMM'] - tfl_df['EntTimeHHMM']
# Grouping By Day of the Week and Calculating the Average Journey Duration of each day
average_journey_times = tfl_df.groupby('downo')['JourneyDuration'].mean()
```

```python
print(average_journey_times)
```

```python
average_journey_times_minutes = average_journey_times.dt.total_seconds() / 60
```

```python
print(average_journey_times_minutes)
```

```python
# Handling negative results (Overnight Journeys)
# Finding rows that have negative journey duration and adding 1 day to account for overnight
tfl_df.loc[tfl_df['JourneyDuration'] < pd.Timedelta(0), 'JourneyDuration'] += pd.Timedelta(days=1)
```

```python
average_journey_times = tfl_df.groupby('downo')['JourneyDuration'].mean()
average_journey_times_minutes = average_journey_times.dt.total_seconds() / 60
print(average_journey_times_minutes)
```

```python
sorted_tfl_df = tfl_df.sort_values(by='downo', ascending=True)
daysOfTheWeek = sorted_tfl_df['daytype'].unique()
print(daysOfTheWeek)
```

```python
import matplotlib.pyplot as plt

# Creating a bar chart for Journey Duration
plt.figure(figsize=(10, 6))
plt.bar(daysOfTheWeek, average_journey_times_minutes.values, color = 'dodgerblue')
plt.ylim(min(average_journey_times_minutes) - 10, max(average_journey_times_minutes) + 10)
plt.xlabel("Day of the Week")
plt.ylabel("Average Journey Duration (minutes)")
plt.title("Average Journey Duration by Day of the Week (Nov 2009)")
plt.xticks(rotation=45, ha='right') #
overall_average = average_journey_times_minutes.mean()
plt.axhline(y=overall_average, color='dimgray', linestyle='-', label='Overall Average of The Week', linewidth=5)
plt.tight_layout()
plt.legend()
plt.show()
```

###**Insight 2:** Which start stations have the highest number of train/tubejourneys during peak hours?


```python
# Filtering journeys during peak times
morning_peak_mask = (tfl_df['EntTimeHHMM'] >= pd.to_timedelta('06:30:00')) & (tfl_df['EntTimeHHMM'] <= pd.to_timedelta('09:30:00'))
evening_peak_mask = (tfl_df['EntTimeHHMM'] >= pd.to_timedelta('16:00:00')) & (tfl_df['EntTimeHHMM'] <= pd.to_timedelta('19:00:00'))
peak_time_mask = morning_peak_mask | evening_peak_mask
morning_peak_time_journeys = tfl_df[morning_peak_mask]
evening_peak_time_journeys = tfl_df[evening_peak_mask]
peak_time_journeys = tfl_df[peak_time_mask]
morning_start_station_counts = morning_peak_time_journeys['StartStn'].value_counts()
evening_start_station_counts = evening_peak_time_journeys['StartStn'].value_counts()
start_station_counts = peak_time_journeys['StartStn'].value_counts()
morning_top_5_stations = morning_start_station_counts.head(6)
evening_top_5_stations = evening_start_station_counts.head(6)
top_5_stations = start_station_counts.head(6)
# As we don't want bus journeys
morning_top_5_stations = morning_top_5_stations[1:]
evening_top_5_stations = evening_top_5_stations[1:]
top_5_stations = top_5_stations[1:]
print(morning_top_5_stations)
print(evening_top_5_stations)
print(top_5_stations)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(morning_top_5_stations.index, morning_top_5_stations.values, color = 'dodgerblue')
plt.ylim(min(morning_top_5_stations) - 1000, max(morning_top_5_stations) + 1000)
plt.xlabel("Start Station")
plt.ylabel("Number of Journeys")
plt.title("Top 5 Start Stations with Most Journeys during Morning Peak Times")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(evening_top_5_stations.index, evening_top_5_stations.values, color = 'dodgerblue')
plt.ylim(min(evening_top_5_stations) - 1000, max(evening_top_5_stations) + 1000)
plt.xlabel("Start Station")
plt.ylabel("Number of Journeys")
plt.title("Top 5 Start Stations with Most Journeys during Evening Peak Times")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_5_stations.index, top_5_stations.values, color = 'dodgerblue')
plt.ylim(min(top_5_stations) - 1000, max(top_5_stations) + 1000)
plt.xlabel("Start Station")
plt.ylabel("Number of Journeys")
plt.title("Top 5 Start Stations with Most Journeys during Peak Times Overall")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

####**Answer**
*   Commuting Patterns and Major Hubs

Morning Peak: The top stations during the morning peak that commuters started their journey in November 2009 in a particular week from most busiest respectively were: London Bridge, Brixton, Victoria, Startford and Liverpool Street. This highlights that commuters not only start from residential zones or suburban areas, but also London terminal hubs from national rail services if they come from South East England or a part of London that isn't connected well by TfL transport links.

Evening Peak: The stations dominating the evening peak in order of business were: Oxford Circus, Liverpool Street, Canary Wharf, London Bridge, Bank. These stations were business districts or areas with high employment density, representing where most commuters end their workday.

Overall Peak: A combination of both peaks shows which stations are consistently the busiest, reflecting their importance in the TfL and they were in order: London Bridge, Liverpool Street, Victoria, Oxford Circus, Canary Wharf
